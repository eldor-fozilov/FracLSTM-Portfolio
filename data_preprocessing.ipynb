```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#!pip install fracdiff
from fracdiff.sklearn import FracdiffStat
```

```python
stock_data = pd.read_csv("data/selected_stocks.csv", header = [0,1], skiprows = [2])
stock_data.head()
```

```python
stock_data.set_index([stock_data.columns[0]], inplace = True)
stock_data.index.name = "Date"
stock_data.head()
```

```python
stock_returns = np.log(stock_data['Adj Close']) - np.log(stock_data['Adj Close'].shift(1))
stock_returns.dropna(inplace = True)
```

```python
stock_returns
```

```python
stock_returns.to_csv("stock_returns.csv")
```

```python
log_stock_prices = np.log1p(stock_data['Adj Close'])
log_stock_prices
```

```python
f = FracdiffStat()

stock_tickers = list(stock_data['Adj Close'].columns)

```

```python
f = FracdiffStat()

stock_tickers = list(stock_data['Adj Close'].columns)

frac_adj_close = f.fit(stock_data['Adj Close'])
df_frac_order = pd.DataFrame()
df_frac_order['Adj Close'] = frac_adj_close.d_

X_frac_adj_close = frac_adj_close.transform(stock_data['Adj Close'])
df_frac_adj_close = pd.DataFrame(X_frac_adj_close, columns = stock_tickers, index=stock_data.index)


frac_log_adj_close = f.fit(log_stock_prices)
df_log_frac_order = pd.DataFrame()
df_log_frac_order['Adj Close'] = frac_log_adj_close.d_

X_frac_log_adj_close = frac_log_adj_close.transform(log_stock_prices)
df_frac_log_adj_close = pd.DataFrame(X_frac_log_adj_close, columns = stock_tickers, index=stock_data.index)


frac_volume = f.fit(stock_data['Volume'])
df_frac_order['Volume'] = frac_volume.d_

X_frac_volume = frac_volume.transform(stock_data['Volume'])
df_frac_volume = pd.DataFrame(X_frac_volume, columns = stock_tickers, index=stock_data.index)

df_frac_order.index = stock_tickers
df_log_frac_order.index = stock_tickers
```
