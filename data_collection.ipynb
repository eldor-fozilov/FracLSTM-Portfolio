```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import yfinance as yf
```

```python
df = pd.read_csv('S&P 500 Historical Components.csv', index_col = 'date')
df.index = pd.to_datetime(df.index)
df
```

```python
# Convert ticker column from csv to list, then sort.
df['tickers'] = df['tickers'].apply(lambda x: sorted(x.split(',')))
df
```

```python
tickers_list = df[df.index.year == 2010].iloc[0]['tickers'] # sample from year 2010
```

```python
len(tickers_list)
```

```python
data = yf.download(tickers_list,'2000-01-01')
```

```python
data = data[['Adj Close', 'Volume']]
data.head()
```

```python
data.index = pd.to_datetime(data.index)
```

```python
df_cleaned = data.dropna(axis = 1)
df_cleaned.head()
```

```python
df_cleaned.shape[1] // 2 # number of stocks
```

```python
df_cleaned.to_csv("selected_stocks.csv")
```
